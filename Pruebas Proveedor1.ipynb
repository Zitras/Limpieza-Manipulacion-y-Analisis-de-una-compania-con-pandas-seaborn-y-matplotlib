```python
# elimino las filas que hay 0 unidades y 0 price

proveedor1 = proveedor1.loc[~(proveedor1[['Price', 'Units']] == 0.00).all(axis=1)]

# Reiniciar los índices del DataFrame
proveedor1 = proveedor1.reset_index(drop=True)
proveedor1
```

```python
#aqui podemos ver las unidades que se vende por dia

proveedor1_grouped = proveedor1.groupby(['Sku', 'Date']).agg({'Units': 'sum', 'Price': 'sum'}).reset_index()

# Mostrar el DataFrame resultante
proveedor1_grouped
```
