```python
import pandas as pd

proveedor2 = pd.read_csv('/Users/Paco/Desktop/T.Grupo/CSV/sellout-proveedor2.csv')
proveedor2 
```

```python
proveedor2_clean = proveedor2.dropna()
proveedor2_clean
```

```python
proveedor2_sorted = proveedor2.sort_values(by='Sku')
proveedor2_sorted
```

```python
proveedor2 = proveedor2.loc[~(proveedor2[['Pieces']] == 0.00).all(axis=1)]

# Reiniciar los índices del DataFrame
proveedor2 = proveedor2.reset_index(drop=True)
proveedor2
```

```python
skus = proveedor2['Sku'].unique()
print(skus)
print(len(skus))
```

```python
df_pivoted = proveedor2.pivot_table(index='Sku', columns='Date', values=['Pieces', 'Price', 'Store_Id'], aggfunc='first')

# Reiniciar los índices del DataFrame
df_pivoted = df_pivoted.reset_index()

# Mostrar el nuevo conjunto de datos pivotado
print(df_pivoted)

```
